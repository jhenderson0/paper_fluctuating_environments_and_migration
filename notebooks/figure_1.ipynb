## Analysis of two-point inequality in immune repertoire data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import random_growth_src.lib as lib

plt.style.use("../project_style.mplstyle")
%config InlineBackend.figure_format = 'retina'
```

```python
df_true = pd.read_csv('../data/tablo/clone_counts_donor_7.csv.gz')
df_well_mixed = pd.read_csv('../data/tablo/well_mixed_clone_counts_donor_7.csv.gz')
```

```python
fig, ax = plt.subplots(figsize=(4, 3.2))

bins = np.linspace(-2.5, 2.5, 20)
ax.hist(lib.get_empirical_log_ratios(df_true, threshold=50), density=True, bins=bins, alpha=0.2, color='black', label='data')
ax.hist(lib.get_empirical_log_ratios(df_well_mixed, threshold=50), density=True, histtype='step', bins=bins, label='well-mixed', linewidth=2, color='black', linestyle='--')
ax.set_yscale('log')

ax.set_xlabel(r'$\log_{10}(C^{\mathrm{obs}}_{\mathrm{blood}}/C^{\mathrm{obs}}_{\mathrm{tonsil}})$')
ax.set_ylabel('Density')
ax.legend(edgecolor='black', loc='upper right',bbox_to_anchor=(1.05, 1))
plt.savefig('../figures/fig_1.svg', bbox_inches='tight', dpi=200, transparent=True)
#plt.savefig('../figures/tablo_densities.pdf', bbox_inches='tight', dpi=200, transparent=True)
```
