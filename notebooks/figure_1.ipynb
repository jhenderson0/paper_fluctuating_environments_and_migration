```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial
import seaborn as sns
import matplotlib.colors as colors
import matplotlib.cm as cm

import random_growth_src.lib as lib

plt.style.use("../project_style.mplstyle")
%config InlineBackend.figure_format = 'retina'
```

## Figure 1

### Figure 1a and b)

```python
fig, ax = plt.subplots(1, 2, figsize=(11, 5), sharey=True, sharex=True)

x_vals = np.linspace(-8, 8, 500)

ratios = [5, 1, 0.01]
ax[0].set_prop_cycle(color=sns.color_palette("Greys_r", len(ratios)))
ax[1].set_prop_cycle(color=sns.color_palette("Greys_r", len(ratios)))
for ratio in ratios:
    ax[0].plot(x_vals, lib.two_patch_steady_state_density_x_white_noise(x_vals, ratio, 1), linewidth=2, alpha=1)
    ax[1].plot(x_vals, lib.mean_field_steady_state_density_x_white_noise(x_vals, ratio, 1), linewidth=2, alpha=1)

ax[0].set_xlim([-8, 8])
ax[0].set_ylim([0, 0.7])

ax[0].text(0.5, 0.6, rf"$M / D = {ratios[0]}$", fontsize=15, weight='bold', c='black')
ax[0].text(1.5, 0.2, rf"$M / D = {ratios[1]}$", fontsize=15, weight='bold', c='black')
ax[0].text(4, 0.1, rf"$M / D = {ratios[2]}$", fontsize=15, weight='bold', c='black')

ax[1].text(0.5, 0.6, rf"$M / D = {ratios[0]}$", fontsize=15, weight='bold', c='black')
ax[1].text(1.5, 0.2, rf"$M / D = {ratios[1]}$", fontsize=15, weight='bold', c='black')
ax[1].text(4, 0.1, rf"$M / D = {ratios[2]}$", fontsize=15, weight='bold', c='black')

ax[0].set_xlabel("$x$", fontsize=25)
ax[0].set_title(r'$N = 2$', fontsize=20)
ax[1].set_title(r'$N \rightarrow \infty $', fontsize=20)

ax[1].set_xlabel("$x$", fontsize=25)
ax[0].set_ylabel(r"$\rho(x)$", fontsize=25);
ax[1].set_ylabel(r"$\rho_{\mathrm{mf}}(x)$", fontsize=25)

ax[0].tick_params(axis="x", which="major", labelsize=20)
ax[1].tick_params(axis="x", which="major", labelsize=20)

ax[0].tick_params(axis="y", which="major", labelsize=20)
ax[1].tick_params(axis="y", which="major", labelsize=20)

#plt.savefig('../figures/fig_1ab.svg', bbox_inches='tight', dpi=200, transparent=True)
plt.savefig('../figures/white_noise_densities.svg', bbox_inches='tight', dpi=200, transparent=True);
```
